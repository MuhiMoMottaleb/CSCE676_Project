# (D) Exploratory Data Analysis (Selected Dataset Only)

## 1. Setup and Loading

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", context="notebook")
plt.rcParams["figure.figsize"] = (10, 6)

# Load data (adjust path as needed)
path = '/content/drive/MyDrive/CSCE676/Project/creditcard.csv'
df = pd.read_csv(path)

print("Shape:", df.shape)
df.head()
```

## 2. Data Info, Missing Values, and Summary Stats

```python
print(df.info())
print("\n")
```

```python
print("========Check for missing values========")
print(df.isna().sum()) # Check missing values in numerical table printout

print("\nHeatmap for checking missing values")
sns.set()
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis') # Plotting missing values (if there are any) in a heatmap.
```

```python
# Transposing the describe function to make it more readable
df.describe().T
```

## 3. Checking Distribution for the Class Variable/Column

```python
class_counts = df['Class'].value_counts()
print(class_counts)
fraud_rate = class_counts[1] / class_counts.sum() * 100
print(f"Fraud rate: {fraud_rate:.4f}%")

plt.figure()
sns.countplot(x='Class', data=df)
plt.title("Class Distribution (0 = Non-fraud, 1 = Fraud)")
plt.show()
```

- Non-fraud (Class = 0): 284,315+
- Fraud (Class = 1): 492
- Fraud rate ~ 0.17%
     
This extreme imbalance is the defining characteristic of the dataset and strongly influences model choice and evaluation.

## 4. Transaction Amount Analysis

```python
print(df['Amount'].describe())

plt.figure()
sns.histplot(df['Amount'], bins=100, kde=False)
plt.yscale("log")
plt.title("Transaction Amount Distribution (log y-axis)")
plt.xlabel("Amount")
plt.show()
```

- `Amount` is heavily right-skewed.
- Many low-value transactions; a long tail of large amounts.
     

```python
df['log_amount'] = np.log1p(df['Amount'])

plt.figure()
sns.histplot(df['log_amount'], bins=100)
plt.title("Log-Transformed Transaction Amount (log1p)")
plt.xlabel("log(Amount + 1)")
plt.show()
```

- Log-transform makes the distribution more symmetric and often improves model performance.

```python
plt.figure()
sns.histplot(data=df, x='Amount', hue='Class', bins=100,
             element='step', stat='density', common_norm=False)
plt.yscale("log")
plt.title("Transaction Amount by Class (log y-axis)")
plt.xlabel("Amount")
plt.show()

plt.figure()
sns.boxplot(x='Class', y='Amount', data=df)
plt.yscale("log")
plt.title("Transaction Amount by Class (Boxplot, log y-axis)")
plt.show()
```

- Fraudulent transactions are spread across different amount ranges, sometimes with a slightly different distribution than non-fraud.
- Amount alone does not separate classes perfectly, but it carries useful signal.
     

## 5. Time Data Structure Analysis

```python
print(df['Time'].describe())
```

```python
# Group into 1-hour bins
bin_size = 3600  # seconds
df['Time_bin'] = (df['Time'] // bin_size).astype(int)

txn_per_bin = df.groupby('Time_bin')['Class'].count()
fraud_per_bin = df.groupby('Time_bin')['Class'].sum()

plt.figure()
txn_per_bin.plot()
plt.title("Number of Transactions per Time Bin (1 hour)")
plt.xlabel("Time Bin (hours since first transaction)")
plt.ylabel("Num Transactions")
plt.show()

plt.figure()
(fraud_per_bin / txn_per_bin).plot()
plt.title("Fraud Rate per Time Bin (1 hour)")
plt.xlabel("Time Bin (hours since first transaction)")
plt.ylabel("Fraud Rate")
plt.show()
```

- Transaction volume fluctuates by time of day (peaks and troughs).
- Fraud rate is noisy due to few frauds; Some bins may have relatively higher or lower fraud proportion.

```python
df_sorted = df.sort_values('Time')
time_diff = df_sorted['Time'].diff().dropna()

print(time_diff.describe())

plt.figure()
sns.histplot(time_diff, bins=100)
plt.yscale("log")
plt.title("Time Gaps Between Consecutive Transactions (log y-axis)")
plt.xlabel("Seconds between transactions")
plt.show()
```

- Many small time gaps -> data bursts noticed; some larger gaps (off-peak times).
- Indicates potential value in viewing this as a streaming / time-aware problem, not purely Indendently Identically Distributed (i.i.d).
     

```python
plt.figure()
sns.kdeplot(data=df, x='Time', hue='Class', common_norm=False)
plt.title("Time Distribution by Class")
plt.xlabel("Time (seconds since first transaction)")
plt.show()
```

Frauds may show mild concentration at certain periods (but patterns are subtle).

## 6. PCA Feature Distribution By Class

```python
features_to_inspect = ['V4', 'V8', 'V12', 'V16', 'V20', 'V24']

for col in features_to_inspect:
    print(f"\n{col} summary:")
    print(df[col].describe())

    plt.figure()
    sns.kdeplot(data=df, x=col, hue='Class', common_norm=False, fill=True, alpha=0.5)
    plt.title(f"Distribution of {col} by Class")
    plt.show()
```

- It should be noted that most of the components have the fraud (class = 1) distribution to have sider variance in comparison to non-fraud (class = 0), except for `V24` component.
- These features are likely important for classification, even though they have no human-interpretable meaning (due to protecting consumer data)

## 7. Correlation Structure Analysis

```python
corr_matrix = df.drop(columns=['Time_bin']).corr()
# The correlation is done in repect to (i.r.t) to Class component/variable
corr_with_class = corr_matrix['Class'].sort_values(ascending=False)

print("Top correlations with Time:")
print(corr_with_class.head(15))
print("\nLowest correlations with Time:")
print(corr_with_class.tail(15))
```

```python
top_corr_features = corr_with_class.abs().sort_values(ascending=False).head(10).index

plt.figure(figsize=(8, 6))
sns.heatmap(df[top_corr_features].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap of Top Features (by Correlation with Class component)")
plt.show()
```

- Some PCA features show non-trivial correlation with Class, both positively and negatively.
- `Amount` and `Time` often have low direct linear correlation with `Class` but can still be useful in non-linear models.
     

## 8. Summary of EDA

- The dataset is clean (no missing values) and fully numeric but heavily imbalanced with fraud making up about 0.17% of transactions.
- Amount is strongly right-skewed; log transformation is recommended. Fraudulent transactions span a range of amounts, with some differences in distribution compared to non-fraud.
- Time-related analysis reveals bursty transaction patterns and varying fraud rates across time bins, suggesting a potential streaming perspective.
- Several PCA-derived features exhibit clear distributional differences between fraud and non-fraud, indicating useful signal for classification despite their anonymized nature.
- Correlation analysis highlights which features are most associated with the fraud label, helping prioritize features and giving intuition for modeling.

# (E) Initial Insights and Direction

## Observations
- The data is well-structured and ready for modeling without extensive cleaning, but interpretability is limited because the main features are anonymized PCA components.
- The dataset is highly imbalanced.  
- A naive classifier that predicts “non-fraud” for every transaction would still achieve >99.8% accuracy, but detect no actual fraud.  
- Amount is likely a useful feature but not sufficient alone.  
- Scaling or log-transforming Amount improves numerical stability and may help many models.  
- There may be “sweet spots” of fraud activity in particular amount ranges, which could inform feature engineering or rule-based checks.
- The data has a natural temporal/streaming structure, not purely IID.  
- In a real setting, models would need to detect fraud in near-real time as transactions appear.  
- Time-based features (e.g., time-of-day, bin counts, recent transaction rates) could be valuable.
- No missing data; all numeric and already standardized to some extent.
- No obvious duplicates at the row level (though this should be confirmed if needed).
- Outliers in Amount and some PCA components exist but are expected in real financial data.    

## Hypothesis
- Hypothesis 1: Models that explicitly handle class imbalance (e.g., class-weighted classifiers or resampling techniques) will achieve significantly higher recall for fraud cases at comparable precision, compared to naïve models trained without imbalance handling.
- Hypothesis 2: Unsupervised or semi-supervised anomaly detection methods (e.g., Isolation Forest, One-Class SVM, autoencoders trained on non-fraud only) can achieve comparable or superior performance to supervised classifiers, especially in terms of recall for the rare fraud class.
- Hypothesis 3: Incorporating time-based features (e.g., time-of-day, recent transaction counts or amounts) will improve fraud detection performance compared to models that ignore temporal information, suggesting that fraud patterns are partially time-dependent.
- Hypothesis 4: Tuning the decision threshold and training models with cost-sensitive objectives (higher penalty for false negatives than false positives) will yield lower expected financial loss, even if raw metrics such as overall accuracy decrease.
